```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.sex.map({"Male":1, "Female":2})
##map은 series에서만 사용가능!!
##딕셔너리 써도 되고 함수를 사용해도 된다!
```

```python
tips.sex.map(lambda x: 1 if x=='Male' else 2)
##함수가 가능!!
## 판다스는 랭글링 툴!! 그래서 쉽게 바꿀 수 있어야 한다. 시리즈 별로 쉽게!!
## 판다스는 함수기반이기 때문에 빠르다~~
```

```python
tips.sex.apply(lambda x: 1 if x=='Male' else 2)
##apply는 Series + DF 다 가능
```

```python
tips.apply(lambda x: x['total_bill']+x["tip"], axis=1)
```

```python
tips[["total_bill","tip"]].apply(sum)
```

```python
tips[["total_bill","tip"]].apply(lambda x : sum(x))
```

```python
tips[["total_bill","tip"]].applymap(lambda x : x+1)
```

판다스의 apply 3총사!!
1. map -Series
2. apply - Series, DF
3. applymap - Series, DF

```python
tips['ratio'] = tips.apply(lambda x:x['tip']/x['total_bill'], axis=1)
```
